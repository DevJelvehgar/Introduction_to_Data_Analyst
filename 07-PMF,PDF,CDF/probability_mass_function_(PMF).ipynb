## Probability Mass Function (PMF):

Used for discrete random variables.<br>
Represents the probability of each possible value.<br>
Example: Rolling a dice, where each face has a probability of 1/6

#### Advantages
1. **Clarity:** Provides a clear representation of probabilities for discrete outcomes.
2. **Simplicity:** Easy to compute and interpret for small datasets.
3. **Foundation for Other Metrics:** Forms the basis for calculating expectations, variances, and other statistical properties.


#### Disadvantages
1. **Limited to Discrete Variables:** Cannot be applied to continuous data.
2. **High Dimensionality:** Becomes unwieldy for variables with a large number of possible values.
3. **Sparse Data:** For variables with many possible values, probabilities can become very small and harder to interpret.


#### When to Use PMF
Counting occurrences (e.g., number of defective items in a batch).<br>
Modeling binary outcomes (e.g., coin tosses).<br>
Events with a fixed number of trials and specific probabilities (e.g., dice rolls).

#### Python Code : PMF of a Dice Roll

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import randint

# Define the random variable (dice roll: outcomes 1 to 6)
outcomes = np.arange(1, 7)
pmf = randint.pmf(outcomes, 1, 7)  # PMF of rolling a fair six-sided die

# Plotting the PMF
plt.figure(figsize=(8, 5))
plt.stem(outcomes, pmf, basefmt=" ")
plt.title("PMF of Rolling a Six-Sided Die")
plt.xlabel("Outcome")
plt.ylabel("Probability")
plt.ylim(0, 0.25)
plt.xticks(outcomes)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# Display PMF Values
for outcome, prob in zip(outcomes, pmf):
    print(f"P(X = {outcome}) = {prob:.2f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import randint

# Define the random variable (dice roll: outcomes 1 to 6)
outcomes = np.arange(1, 7)
pmf = randint.pmf(outcomes, 1, 7)  # PMF of rolling a six-sided die
cdf = np.cumsum(pmf)  # Calculate CDF as cumulative sum of PMF

# Plotting the PMF and CDF
plt.figure(figsize=(12, 6))

# PMF Plot
plt.subplot(1, 2, 1)
plt.stem(outcomes, pmf, basefmt=" ", label="PMF")
plt.title("PMF of Rolling a Six-Sided Die")
plt.xlabel("Outcome")
plt.ylabel("Probability")
plt.xticks(outcomes)
plt.ylim(0, 0.25)
plt.legend()
plt.grid(axis="y", linestyle="--", alpha=0.7)

# CDF Plot
plt.subplot(1, 2, 2)
plt.step(outcomes, cdf, where="post", label="CDF", color="orange")
plt.scatter(outcomes, cdf, color="red")  # Highlight the CDF values at each outcome
plt.title("CDF of Rolling a Six-Sided Die")
plt.xlabel("Outcome")
plt.ylabel("Cumulative Probability")
plt.xticks(outcomes)
plt.ylim(0, 1.1)
plt.legend()
plt.grid(axis="y", linestyle="--", alpha=0.7)

plt.tight_layout()
plt.show()

# Display CDF Values
for outcome, prob in zip(outcomes, cdf):
    print(f"F(X <= {outcome}) = {prob:.2f}")
```
