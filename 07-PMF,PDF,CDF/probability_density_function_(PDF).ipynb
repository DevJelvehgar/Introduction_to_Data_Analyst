## What is the Probability Density Function (PDF)?
A Probability Density Function (PDF) describes the likelihood of a continuous random variable taking on a specific value. Unlike discrete probabilities, the PDF does not directly give probabilities but represents the density of probabilities over an interval.

#### Key Features:
The area under the PDF curve over an interval gives the probability for that interval.<br>
For continuous variables, the probability at an exact point is 0. Instead, we calculate probabilities over intervals.<br>

#### Example:
In a normal distribution, the PDF tells us how the values are distributed around the mean, showing the likelihood of values near the mean compared to those further away.

#### Advantages
1. Smooth Representation: Provides a smooth curve for modeling continuous random variables.
2. Versatile: Can represent various distributions like normal, exponential, and uniform.
3. Interpretable: Useful in visualizing and understanding the distribution of data.

#### Disadvantages
1. Complexity: Difficult to compute probabilities for non-standard distributions.
2. Requires Integration: Calculating probabilities over intervals requires numerical or analytical integration.
3. Not Applicable to Discrete Data: Only works for continuous variables.


#### When to Use PDF
To model and analyze continuous data (e.g., heights, weights, or time).<br>
When probabilities over intervals are needed (e.g., the likelihood of temperatures between 20°C and 25°C).<br>
In probability theory and statistical modeling (e.g., hypothesis testing).


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Define the PDF: Normal Distribution
mean = 0  # Mean of the distribution
std_dev = 1  # Standard deviation
x = np.linspace(-4, 4, 1000)  # Range of x values
pdf = norm.pdf(x, loc=mean, scale=std_dev)  # Compute PDF

# Plot the PDF
plt.figure(figsize=(8, 5))
plt.plot(x, pdf, color='blue', label=f'N({mean}, {std_dev}²)')
plt.fill_between(x, pdf, alpha=0.3, color='skyblue')  # Shade under the curve
plt.title("Probability Density Function (PDF) - Normal Distribution")
plt.xlabel("Value")
plt.ylabel("Density")
plt.axvline(mean, color='red', linestyle='--', label='Mean')
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Define the PDF and CDF: Normal Distribution
mean = 0  # Mean of the distribution
std_dev = 1  # Standard deviation
x = np.linspace(-4, 4, 1000)  # Range of x values

pdf = norm.pdf(x, loc=mean, scale=std_dev)  # Compute PDF
cdf = norm.cdf(x, loc=mean, scale=std_dev)  # Compute CDF

# Plot the PDF
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.plot(x, pdf, color='blue', label=f'PDF (N({mean}, {std_dev}²))')
plt.fill_between(x, pdf, alpha=0.3, color='skyblue')  # Shade under the curve
plt.title("Probability Density Function (PDF)")
plt.xlabel("Value")
plt.ylabel("Density")
plt.axvline(mean, color='red', linestyle='--', label='Mean')
plt.legend()
plt.grid(alpha=0.3)

# Plot the CDF
plt.subplot(1, 2, 2)
plt.plot(x, cdf, color='green', label='CDF')
plt.title("Cumulative Distribution Function (CDF)")
plt.xlabel("Value")
plt.ylabel("Cumulative Probability")
plt.axhline(0.5, color='red', linestyle='--', label='Median (0.5)')
plt.legend()
plt.grid(alpha=0.3)

plt.tight_layout()
plt.show()
```
