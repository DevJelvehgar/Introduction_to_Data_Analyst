# What is Descriptive Statistics?
Descriptive Statistics refers to the statistical methods used to **summarize**, **organize**, and **simplify** data. It provides an overview of the main features of a data set, making it easier to understand and communicate patterns or trends. This involves measures of central tendency, variability, and distribution shape.

## Types of Descriptive Statistics

#### 1.Measures of Central Tendency: These summarize the center of a data set
        1.1 Mean (average)
        1.2 Median (middle value when data is ordered)
        1.3 Mode (most frequently occurring value)

#### 2.Measures of Dispersion (Variability): These describe the spread or variation in data
        2.1 Range (difference between maximum and minimum values)
        2.2 Variance (average of squared deviations from the mean)
        2.3 Standard Deviation (square root of variance)
#### 3.Measures of Shape:
        3.1 Skewness: Indicates asymmetry of the data distribution.
        3.2 Kurtosis: Measures the "tailedness" of the data distribution
        
#### 4.Frequency Distributions: Tabular or graphical representation of data.
        4.1 Histograms
        4.2 Frequency tables
        4.3 Bar charts

#### 5.Summary Statistics
        5.1 five number summary

## Advantages and Disadvantages
#### Advantages:
1. Simplicity: Provides a straightforward summary of the data.
2. Quick Insights: Helps detect trends and patterns at a glance.
3. Data Comparison: Enables easy comparison across datasets.
#### Disadvantages:
1. No Cause-and-Effect: Cannot infer relationships or causality.
2. Data Oversimplification: May hide important nuances or outliers.
3. Limited Use for Complex Analysis: Not suitable for hypothesis testing or making predictions.

## When to Use Descriptive Statistics?
1. Initial Data Exploration: To understand data before conducting further analyses.
2. Summarizing Large Data Sets: When presenting data to stakeholders or in reports.
3. Detecting Anomalies: To identify outliers or unusual patterns.


## Example Scenario
Suppose you have a dataset containing the ages of participants in a survey. 
Descriptive statistics can help summarize this data by providing the mean age, range, and identifying the age that appears most frequently.

```python
# python code
import numpy as np
import pandas as pd
from scipy.stats import skew, kurtosis

# Sample dataset: Ages of participants
data = [25, 30, 35, 40, 45, 50, 55, 60, 65, 70, 30, 35, 35, 45]

# Convert to a Pandas DataFrame for better handling
df = pd.DataFrame(data, columns=['Age'])

# Descriptive statistics
mean = df['Age'].mean()
median = df['Age'].median()
mode = df['Age'].mode()[0]  # Mode might return multiple values
range_value = df['Age'].max() - df['Age'].min()
variance = df['Age'].var()
std_dev = df['Age'].std()
skewness = skew(df['Age'])
kurt = kurtosis(df['Age'])

# Print results
print("Descriptive Statistics:")
print(f"Mean: {mean:.2f}")
print(f"Median: {median:.2f}")
print(f"Mode: {mode:.2f}")
print(f"Range: {range_value}")
print(f"Variance: {variance:.2f}")
print(f"Standard Deviation: {std_dev:.2f}")
print(f"Skewness: {skewness:.2f}")
print(f"Kurtosis: {kurt:.2f}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Visualization settings
plt.style.use('ggplot')

# Histogram to visualize the distribution of ages
plt.figure(figsize=(10, 4))
sns.histplot(df['Age'], bins=10, kde=True, color='orange', edgecolor='black')
plt.axvline(mean, color='red', linestyle='dashed', linewidth=2, label=f'Mean: {mean:.2f}')
plt.axvline(median, color='green', linestyle='dashed', linewidth=2, label=f'Median: {median}')
plt.axvline(mode, color='blue', linestyle='dashed', linewidth=2, label=f'Mode: {mode}')
plt.title("Age Distribution with Mean, Median, and Mode")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.legend()
plt.show()
```

```python
## five number summary
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Sample dataset: Test scores of students
data = [55, 60, 61, 62, 65, 68, 70, 72, 75, 78, 80, 82, 85, 87, 90, 95, 99]

# Convert to a Pandas Series
data_series = pd.Series(data)

# Calculate the Five-Number Summary
minimum = data_series.min()
q1 = data_series.quantile(0.25)  # 25th percentile
median = data_series.median()  # 50th percentile
q3 = data_series.quantile(0.75)  # 75th percentile
maximum = data_series.max()

# Display the results
print("Five-Number Summary:")
print(f"Minimum: {minimum}")
print(f"Q1 (25th Percentile): {q1}")
print(f"Median (Q2): {median}")
print(f"Q3 (75th Percentile): {q3}")
print(f"Maximum: {maximum}")

# Boxplot to visualize the Five-Number Summary
plt.figure(figsize=(8, 4))
sns.boxplot(x=data_series, color='skyblue')
plt.title("Boxplot of Student's Marks")
plt.xlabel("Student's Marks")
plt.show()
```
