# **Binomial Distribution**

The **Binomial Distribution** is a discrete probability distribution that describes the number of successes in a fixed number of independent Bernoulli trials, each with the same probability of success.

---

### **Formula for Binomial Distribution**
The probability mass function (PMF) for a Binomial random variable \( X \) is given by:

$$
P(X = k) = \binom{n}{k} p^k (1-p)^{n-k}
$$

Where:
- \( n \): Total number of trials
- \( k \): Number of successes ( k = 0, 1, ..., n )
- \( p \): Probability of success for a single trial
- Binomial coefficient : $$ ( \binom{n}{k} = \frac{n!}{k!(n-k)!}) $$

---

### **Advantages of Binomial Distribution**
1. **Widely Applicable**: Used in various real-world scenarios, such as quality control, survey analysis, and risk assessment.
2. **Foundation for Probability Models**: Provides a basis for other distributions like the Poisson and Normal distributions.
3. **Simple to Compute**: The PMF and other properties can be easily calculated.

---

### **Disadvantages of Binomial Distribution**
1. **Fixed Number of Trials**: Requires the total number of trials (\(n\)) to be known beforehand.
2. **Independent Trials Assumption**: Assumes each trial is independent, which may not always hold in practice.
3. **Constant Probability**: Assumes the probability of success (\(p\)) remains the same across all trials.

---

### **Properties of Binomial Distribution**

1. **Mean**:
   $$
   \mu = np
   $$

2. **Variance**:
   $$
   \sigma^2 = np(1-p)
   $$

3. **Standard Deviation**:
   $$
   \sigma = \sqrt{np(1-p)}
   $$

4. **Mode**:
   $$
   \text{Mode} = \lfloor (n+1)p \rfloor \quad \text{or} \quad \lceil (n+1)p \rceil
   $$

5. **Median**:
   The median is approximately $$ ( \lfloor np \rfloor ) or ( \lceil np \rceil )$$

---



```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

# Parameters for Binomial Distribution
n = 10  # Total number of trials
p = 0.5  # Probability of success for a single trial

# Create a Binomial random variable
x = np.arange(0, n + 1)  # Possible values of k (0 to n)
pmf = binom.pmf(x, n, p)  # PMF of Binomial distribution
cdf = binom.cdf(x, n, p)  # CDF of Binomial distribution

# Calculate properties
mean = n * p
variance = n * p * (1 - p)
std_dev = np.sqrt(variance)
mode = int((n + 1) * p)  # For simplicity, using the floor value
median = int(n * p)

# Display the results
print(f"Mean: {mean}")
print(f"Variance: {variance}")
print(f"Standard Deviation: {std_dev}")
print(f"Mode: {mode}")
print(f"Median: {median}")

# Plot the PMF and CDF
fig, ax1 = plt.subplots(figsize=(10, 6))

# PMF plot
ax1.bar(x, pmf, color='skyblue', alpha=0.7, label="PMF", width=0.4, align='center')
ax1.set_xlabel("Number of Successes (k)")
ax1.set_ylabel("Probability", color="blue")
ax1.axvline(mean, color='red', linestyle='--', label='Mean')
ax1.legend(loc='upper left')
ax1.set_title("Binomial Distribution: PMF and CDF")
ax1.grid(True)

# CDF plot
ax2 = ax1.twinx()
ax2.plot(x, cdf, color='green', marker='o', label="CDF")
ax2.set_ylabel("Cumulative Probability", color="green")
ax2.legend(loc='upper right')

plt.show()
```
