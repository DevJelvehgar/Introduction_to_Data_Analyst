# **Types of Correlation**

Correlation is a statistical measure that expresses the extent to which two variables are related. Below are the main types of correlation:

---

## **1. Positive Correlation**
- **Definition**: As one variable increases, the other variable also increases.
- **Example**: Height and weight (taller people tend to weigh more).
- **Correlation Coefficient**: Positive values close to +1 (e.g., 0.8, 0.9).

---

## **2. Negative Correlation**
- **Definition**: As one variable increases, the other variable decreases.
- **Example**: Temperature and the amount of hot coffee sold (as temperature rises, coffee sales decrease).
- **Correlation Coefficient**: Negative values close to -1 (e.g., -0.7, -0.8).

---

## **3. Zero (No) Correlation**
- **Definition**: No relationship exists between the two variables; changes in one variable do not affect the other.
- **Example**: Shoe size and intelligence.
- **Correlation Coefficient**: Close to 0 (e.g., -0.1, +0.1).

---

## **4. Linear Correlation**
- **Definition**: The relationship between two variables can be represented by a straight line.
- **Example**: The number of hours studied and exam scores.
- **Methods**: Measured using Pearson’s correlation coefficient.

---

## **5. Non-Linear (Curvilinear) Correlation**
- **Definition**: The relationship between variables follows a curve rather than a straight line.
- **Example**: The stress level vs. performance (performance increases with moderate stress but decreases with excessive stress).
- **Methods**: Typically not measured with Pearson’s coefficient but using Spearman’s or other methods.

---

## **6. Pearson Correlation Coefficient**
- **Definition**: Measures the linear relationship between two continuous variables.
- **Range**: -1 to +1.
  - **+1**: Perfect positive linear correlation.
  - **-1**: Perfect negative linear correlation.
  - **0**: No linear correlation.

---

## **7. Spearman Rank Correlation**
- **Definition**: Measures the relationship between two ranked (ordinal) variables.
- **When to Use**: When data is not normally distributed or has non-linear relationships.
- **Range**: -1 to +1.

---

## **8. Kendall Tau Correlation**
- **Definition**: Measures the strength and direction of association between two ranked variables.
- **When to Use**: For smaller datasets or data with tied ranks.
- **Range**: -1 to +1.

---

## **9. Partial Correlation**
- **Definition**: Measures the correlation between two variables while controlling for the effect of other variables.
- **Example**: Correlation between exercise and weight loss, controlling for calorie intake.

---

## **10. Point-Biserial Correlation**
- **Definition**: Measures the relationship between a continuous variable and a binary variable.
- **Example**: Test scores (continuous) and gender (binary).

---
