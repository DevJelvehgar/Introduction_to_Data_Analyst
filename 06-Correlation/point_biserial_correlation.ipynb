```python
import pandas as pd
import numpy as np
from scipy.stats import pointbiserialr
import matplotlib.pyplot as plt

# Step 1: Create sample data
np.random.seed(42)
data = pd.DataFrame({
    'Test_Score': np.random.normal(75, 10, 100),  # Continuous variable (test scores)
    'Gender': np.random.choice([0, 1], size=100, p=[0.5, 0.5])  # Binary variable (0 = Male, 1 = Female)
})

# Step 2: Compute Point-Biserial Correlation
correlation, p_value = pointbiserialr(data['Gender'], data['Test_Score'])

# Print the results
print(f"Point-Biserial Correlation Coefficient: {correlation:.2f}")
print(f"P-value: {p_value:.4f}")

# Step 3: Plot the data
plt.figure(figsize=(10, 6))
plt.scatter(data['Gender'], data['Test_Score'], alpha=0.7, color='blue')
plt.xticks([0, 1], ['Male', 'Female'])
plt.xlabel('Gender')
plt.ylabel('Test Score')
plt.title(f"Point-Biserial Correlation\nCorrelation Coefficient: {correlation:.2f}, P-value: {p_value:.4f}", fontsize=14)
plt.grid(alpha=0.5)
plt.show()
```
