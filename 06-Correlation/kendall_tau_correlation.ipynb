## Kendall Tau Correlation

$$
\tau = \frac{\text{Number of Concordant Pairs} - \text{Number of Discordant Pairs}}{\frac{n(n-1)}{2}}
$$



$$
\tau = \frac{C - D}{\frac{n(n-1)}{2}}
\
\
\text{Where:}
\begin{align*}
C & = \text{Number of Concordant Pairs} \\
D & = \text{Number of Discordant Pairs} \\
n & = \text{Total number of data points.}
\end{align*} $$


#### When to Use Kendall Tau Correlation:
1. When your data is ordinal or ranked.<br>
2. When you want to analyze monotonic relationships.<br>
3. For small sample sizes or when there are many tied ranks in the data.

```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import kendalltau

# Step 1: Real-world ordinal data
# Customer satisfaction (1 = Very Dissatisfied, 2 = Dissatisfied, 3 = Neutral, 4 = Satisfied, 5 = Very Satisfied)
# Product quality (1 = Poor, 2 = Fair, 3 = Good, 4 = Very Good, 5 = Excellent)

data = pd.DataFrame({
    'Customer_Satisfaction': [4, 5, 3, 4, 2, 1, 3, 5, 4, 3, 2, 4, 5, 3, 1, 2, 4, 5, 4, 3],
    'Product_Quality':       [5, 4, 3, 4, 2, 1, 3, 5, 4, 3, 2, 4, 5, 4, 1, 2, 5, 4, 5, 3]
})

# Step 2: Compute Kendall Tau Correlation
tau, p_value = kendalltau(data['Customer_Satisfaction'], data['Product_Quality'])

# Step 3: Print correlation results
print(f"Kendall Tau Correlation Coefficient: {tau:.2f}")
print(f"P-value: {p_value:.4f}")

# Step 4: Scatter plot visualization
plt.figure(figsize=(8, 6))
plt.scatter(data['Customer_Satisfaction'], data['Product_Quality'], alpha=0.7, color='blue')
plt.title(f"Kendall Tau Correlation\nCorrelation Coefficient: {tau:.2f}, P-value: {p_value:.4f}", fontsize=14)
plt.xlabel('Customer Satisfaction', fontsize=12)
plt.ylabel('Product Quality', fontsize=12)
plt.xticks([1, 2, 3, 4, 5], ['Very Dissatisfied', 'Dissatisfied', 'Neutral', 'Satisfied', 'Very Satisfied'])
plt.yticks([1, 2, 3, 4, 5], ['Poor', 'Fair', 'Good', 'Very Good', 'Excellent'])
plt.grid(alpha=0.5)
plt.show()
```
