# Pearson Correlation Coefficient Formula

The Pearson correlation coefficient (\(r\)) is a measure of the linear relationship between two variables \(X\) and \(Y\). It is calculated using the following formula:

$$
r = \frac{\sum{(X_i - \bar{X})(Y_i - \bar{Y})}}{\sqrt{\sum{(X_i - \bar{X})^2} \cdot \sum{(Y_i - \bar{Y})^2}}}
$$


```python
import numpy as np
import pandas as pd
from scipy.stats import pearsonr
import matplotlib.pyplot as plt

# Generate sample data
np.random.seed(42)
x = np.random.rand(100) * 10  # Variable X: Random values between 0 and 10
y = 2 * x + np.random.normal(loc=0, scale=3, size=100)  # Variable Y: Linearly dependent on X with noise

# Create a DataFrame
data = pd.DataFrame({'X': x, 'Y': y})

# Calculate Pearson Correlation
correlation, p_value = pearsonr(data['X'], data['Y'])
print(f"Pearson Correlation Coefficient: {correlation:.3f}")
print(f"P-value: {p_value:.3f}")

# Plot the data
plt.scatter(data['X'], data['Y'], alpha=0.7, color='blue', label='Data Points')
plt.title("Scatter Plot of X vs. Y")
plt.xlabel('X')
plt.ylabel('Y')
plt.axvline(np.mean(data['X']), color='red', linestyle='--', label='Mean of X')
plt.axhline(np.mean(data['Y']), color='green', linestyle='--', label='Mean of Y')
plt.legend()
plt.grid(True)
plt.show()

# Interpretation
if p_value < 0.05:
    print("The correlation is statistically significant.")
else:
    print("The correlation is not statistically significant.")
```

```python
import numpy as np
from scipy.stats import pearsonr

# Data for Temperature and Clothing
temperature = np.array([5, 10, 15, 20, 25, 30])
clothing = np.array([10, 8, 6, 5, 3, 2])

# Calculate Pearson correlation
correlation, p_value = pearsonr(temperature, clothing)

print(f"Pearson Correlation Coefficient: {correlation}")
# Create a scatter plot
plt.scatter(temperature, clothing, color='blue', label='Data points')

# Adding labels
plt.title('Negative Correlation: Temperature vs Clothing')
plt.xlabel('Temperature (°C)')
plt.ylabel('Clothing (Items worn)')

# Show plot
plt.show()
```

### Outputs:
1. **correlation:**
A value between -1 and +1.<br>
+1: Perfect positive linear relationship.<br>
-1: Perfect negative linear relationship.<br>
0: No linear relationship.<br>

2. **p_value:**
A value between 0 and 1.<br>
Indicates the probability of observing the data assuming there is no correlation (null hypothesis).<br>
Low p-value (< 0.05): Reject the null hypothesis, meaning the correlation is statistically significant.<br>
High p-value (>= 0.05): Fail to reject the null hypothesis, meaning the correlation might not be significant.<br>
