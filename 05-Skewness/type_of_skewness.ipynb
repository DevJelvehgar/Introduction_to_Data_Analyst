## Types of Skewness

1. **Symmetrical Distribution:**<br>
Example: Normal distribution.<br>
Mean = Median = Mode.<br>
Commonly found in naturally occurring data like heights, weights, etc.<br>

2. **Right-Skewed (Positive Skewness):**<br>
Long tail on the right side: mean > median > mode.<br>
Example: Income distribution (many low incomes, few very high incomes).<br>
Mean is influenced by extreme high values, pulling it to the right.<br>

3. **Left-Skewed (Negative Skewness):**<br>
Long tail on the left side.<br>
Tail extends to the left; mean < median < mode.<br>
Example: Exam scores (most students score high, a few score low).<br>
Mean is influenced by extreme low values, pulling it to the left.<br>


### Advantages and Disadvantages of Skewness Types

| **Skewness Type** | **Advantages**                                   | **Disadvantages**                             | **When to Use**                                                |
|--------------------|-------------------------------------------------|-----------------------------------------------|----------------------------------------------------------------|
| **Symmetrical**    | Easy to interpret; Mean is a good measure.       | Rare in real-world data.                      | Use for normally distributed data (e.g., Gaussian models).     |
| **Right-Skewed**   | Highlights rare high-value cases.                | Mean can be misleading due to extreme values. | Use for income, wealth, or sales data.                        |
| **Left-Skewed**    | Emphasizes rare low-value cases.                 | Mean can be misleading due to extreme values. | Use for cases like exam scores or time-to-completion data.     |


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import skew

# Generate data
np.random.seed(42)

# Symmetrical distribution : Mean = loc, Standard-Deviation = scale, Size = random samples to generate
symmetrical = np.random.normal(loc=50, scale=10, size=1000)

# Right-skewed distribution: For the exponential distribution, the mean = scale.
right_skewed = np.random.exponential(scale=10, size=1000)


# Left-skewed distribution
# This subtracts the generated values from 100. 
# Subtracting an exponential distribution flips the tail direction, creating a left-skewed distribution.
left_skewed = 100 - np.random.exponential(scale=10, size=1000)

# Create a DataFrame
data = pd.DataFrame({
    'Symmetrical': symmetrical,
    'Right-Skewed': right_skewed,
    'Left-Skewed': left_skewed
})

# Plot Distributions
fig, axes = plt.subplots(1, 3, figsize=(18, 6))
titles = ['Symmetrical', 'Right-Skewed', 'Left-Skewed']

for ax, col, title in zip(axes, data.columns, titles):
    ax.hist(data[col], bins=30, alpha=0.7, color='skyblue', label=col)
    ax.axvline(data[col].mean(), color='red', linestyle='--', label='Mean')
    ax.axvline(np.median(data[col]), color='green', linestyle='-', label='Median')
    ax.axvline(data[col].round().mode().values[0], color='orange', linestyle=':', label='Mode')
    ax.set_title(title)
    ax.legend()

plt.tight_layout()
plt.show()
```
